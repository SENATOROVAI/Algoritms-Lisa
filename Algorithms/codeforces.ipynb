В кодфорс не используется нотация BigO  а используется время в секундах по ряду причин:


1. Практическая применимость:

Время в секундах более понятно для пользователей, не знакомых с теорией сложности алгоритмов.
Оно напрямую отражает время, которое требуется на выполнение алгоритма, что важно для оценки его практической эффективности.

2. Неоднозначность BigO:

Обозначение BigO не даёт точной оценки времени выполнения алгоритма.
Оно лишь определяет асимптотическое поведение алгоритма при увеличении размера входных данных.

3. Необходимость сравнения:

В задаче требуется сравнить два алгоритма.
Сравнение по времени выполнения в секундах более наглядно и понятно, чем сравнение по обозначениям BigO.

4. Ограниченность BigO:

Обозначение BigO не учитывает такие факторы, как:
Конкретная реализация алгоритма
Аппаратное обеспечение
Операционная система

5. Простота:

Использование времени в секундах проще и понятнее для начинающих программистов.

```python
# https://codeforces.com/problemset/problem/4/A
w = int(input())
if w % 2 == 0 and w > 2:
    print('Yes')
else:
    print('No')
```
